# 3.which airline is the most punctual?

```python
data_merge['total_delay'] = data_merge['arr_delay'] + data_merge['dep_delay']
```

```python
data_merge.groupby(by=['carrier','airline_name']).agg(mean_delay=('total_delay', 'mean')).sort_values(by='mean_delay').reset_index()
```

# 4.what destination is the longest duration

```python
data_merge['total_time'] =  data_merge['arr_time'] - data_merge['dep_time']
```

```python
data_merge[['air_time', 'total_time']]
```

```python
data_merge.groupby(by=['origin', 'dest']).agg(mean_arr=('air_time', 'mean')).sort_values(by='mean_arr', ascending=False).reset_index()
```

# 5.what airline is the worst in terms of delays

```python
data_merge.groupby(by=['carrier','airline_name']).agg(mean_delay=('total_delay', 'mean')).sort_values(by='mean_delay', ascending=False).reset_index()
```

# 6.which airline has the highest capacity of seats?

```python
fligh_tmp = flights[['carrier', 'tailnum']].drop_duplicates()
```

```python
flight_merge = pd.merge(fligh_tmp, planes[['tailnum', 'seats']], how='left')
```

```python
flight_merge.groupby('carrier').agg(total_seats=('seats', 'sum')).sort_values(by='total_seats', ascending=False)
```

# 7.which airplane model is the highest in use and from which manufacturer?

```python
data_merge.groupby(['model', 'manufacturer']).agg(total_seats=('tailnum', 'count')).sort_values(by='total_seats', ascending=False).reset_index()
```
