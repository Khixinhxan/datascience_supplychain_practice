```python
country_pivoted = pd.pivot_table(country_date_qty, index='date', 
                                 columns='Country', values=['Quantity', 'Price'], 
                                 fill_value = 0,
                                 aggfunc={'Quantity': np.sum, 
                                          'Price': np.mean
                                         }).reset_index()
```

```python
destingnation_data = pd.DataFrame({'name': ['mike', 'jonathan', 'Mo', 'Lee', 'Raj', 'Teo'],
                                   'title': ['manager', 'supervisor', 'director', 'associate','assistant', 'sectionhead']})


age_data = pd.DataFrame({'name': ['mike', 'jonathan', 'Mo', 'Lisa', 'Raj', 'Teo'],
                                   'age': [40,50,34,25,60,39] })

```

```python
destingnation_data
```

```python
age_data
```

```python
# left join
pd.merge(destingnation_data, age_data, how='left')
```

```python
# full join
pd.merge(destingnation_data, age_data, how='outer')
```

```python
# inner join
pd.merge(destingnation_data, age_data, how='inner')
```
