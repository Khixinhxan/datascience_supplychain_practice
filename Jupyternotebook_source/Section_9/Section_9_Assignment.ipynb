```python
import pandas as pd
import numpy as np
import inventorize3 as inv
import os 
path  = os.getcwd()
print(path)
```

```python
# 1 Import twenty eleven
```

```python
twenty_eleven = pd.read_csv("{0}/twentyeleven.csv".format(path))
```

```python
twenty_eleven.head()
```

```python
twenty_eleven.columns
```

```python
# 2 Make sure to remove rows that have description NAs
```

```python
twenty_eleven = twenty_eleven.dropna()
```

```python
# 3 manipulate data to have the quantity and revenue per SKU
```

```python
twenty_eleven['Revenue'] = twenty_eleven['Quantity'] * twenty_eleven['Price']
```

```python
twenty_eleven.head()
```

```python
twenty_eleven_group = twenty_eleven.groupby(['Description']).agg(total_sales=('Quantity', 'sum'),
                                                                 total_revenue=('Revenue', 'sum')
                                                                ).reset_index()
```

```python
twenty_eleven_group
```

```python
# 3 Apply the product_mix function
```

```python
a = inv.ABC(twenty_eleven_group[['Description', 'total_sales']])
```

```python
a.Category.value_counts()
```

```python
b = inv.productmix(twenty_eleven_group['Description'], twenty_eleven_group['total_sales'], twenty_eleven_group['total_revenue'])
```

```python
# 4 Manipulate data to have the quantity and revenue per SKY per country
```

```python
twenty_eleven_group = twenty_eleven.groupby(['Country','Description']).agg(total_sales=('Quantity', 'sum'),
                                                                 total_revenue=('Revenue', 'sum')
                                                                ).reset_index()
```

```python
twenty_eleven_group
```

```python
c = inv.productmix_storelevel(twenty_eleven_group['Description'], 
                              twenty_eleven_group['total_sales'],
                              twenty_eleven_group['total_revenue'],
                              twenty_eleven_group['Country']
                             )
```

```python
c
```

```python
c.storeofsku.value_counts()
```

```python
c_group = c.groupby(['storeofsku', 'product_mix']).count().reset_index()
```
